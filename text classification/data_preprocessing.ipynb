```python
from datasets import load_dataset
dataset = load_dataset("naver-clova-ix/cord-v2")
```

```python
import json
```

```python
def preprocess(subset):
  item_names = []
  data = dataset[subset]
  for content in data:
    data_dict = json.loads(content['ground_truth'])
    menu = data_dict['gt_parse']['menu']
    if isinstance(menu, dict):
      try:
        item_names.append(menu['nm'])
      except KeyError:
        print('[INFO] no atribut "nm"')
    elif isinstance(menu, list):
        for item in menu:
          try:
            item_names.append(item['nm'])
          except KeyError:
            print('[INFO] no atribut "nm"')
  return item_names
```

```python
train_data = preprocess('train')
val_data = preprocess('validation')
test_data = preprocess('test')
```

```python
for i, item in enumerate(train_data):
    if not isinstance(item, str):
        print(f"Non-string element found at index {i}: {item}")
```
